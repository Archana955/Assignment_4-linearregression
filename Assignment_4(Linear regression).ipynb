```python
sns.distplot(salary['Salary'], bins = 10, kde = True)
plt.title('Before Transformation')
sns.displot(np.log(salary['Salary']), bins = 10, kde = True)
plt.title('After log Transformation')
sns.displot(np.sqrt(salary['Salary']), bins = 10, kde = True)
plt.title('After Sqrt Transformation')
sns.displot(np.square(salary['Salary']), bins = 10, kde = True)
plt.title('After Square Transformation')
plt.show()
```

# **Model building**

```python
import statsmodels.formula.api as smf
model =smf.ols("Salary~YearsExperience",data=salary).fit()
```

```python
sns.regplot(x="YearsExperience",y="Salary",data=salary);
```

```python
#coefficients
model.params
```

```python
#R squared value
(model.rsquared,model.rsquared_adj)
```

```python
#t & p-values
print(model.tvalues,'\n',model.pvalues)
```

```python
# mean square error
model.mse_resid
```

```python
# root mean square error
np.sqrt(model.mse_resid)
```

# **Log transformation on data**

```python
model1 = smf.ols('np.log(Salary)~np.log(YearsExperience)', data = salary).fit()
```

```python
#coefficients
model1.params
```

```python
#R squared value
(model1.rsquared,model1.rsquared_adj)
```

```python
#t & p-values
print(model1.tvalues,'\n',model1.pvalues)
```

```python
# mean square error
model1.mse_resid
```

# **Square transformation on data**

```python
model2 = smf.ols('np.square(Salary)~np.square(YearsExperience)', data = salary).fit()
```

```python
#coefficients
model2.params
```

```python
#R squared value
(model2.rsquared,model2.rsquared_adj)
```

```python
#t & p-values
print(model2.tvalues,'\n',model2.pvalues)
```

```python
# mean square error
model2.mse_resid
```

# **Sqrt transformation on data**

```python
model3 = smf.ols('np.sqrt(Salary)~np.sqrt(YearsExperience)', data = salary).fit()
```

```python
#coefficients
model3.params
```

```python
#R squared value
(model3.rsquared,model3.rsquared_adj)
```

```python
#t & p-values
print(model3.tvalues,'\n',model3.pvalues)
```

```python
# mean square error
model3.mse_resid
```

# **Comparing different models with respect to the Root Mean Squared Errors**

```python
from sklearn.metrics import mean_squared_error
```

```python
model1_pred_y =np.exp(model1.predict(salary['YearsExperience']))
model2_pred_y =np.sqrt(model2.predict(salary['YearsExperience']))
model3_pred_y =np.square(model3.predict(salary['YearsExperience']))
```

```python
model1_rmse =np.sqrt(mean_squared_error(salary['Salary'], model1_pred_y))
model2_rmse =np.sqrt(mean_squared_error(salary['Salary'], model2_pred_y))
model3_rmse =np.sqrt(mean_squared_error(salary['Salary'], model3_pred_y))
print('model=', np.sqrt(model.mse_resid),'\n' 'model1=', model1_rmse,'\n' 'model2=', model2_rmse,'\n' 'model3=', model3_rmse)
```

```python
rmse = {'model': np.sqrt(model.mse_resid), 'model1': model1_rmse, 'model2': model3_rmse, 'model3' : model3_rmse}
min(rmse, key=rmse.get)
```

# **Therefore,  'model' has the minimum RMSE and highest Adjusted R-squared score. Hence, we will use 'model' to predict our y variable which is Salary**

```python
# predict  Salary for 15.5 and 18.0 years of experience
newdata=pd.Series([15.5,18.0])
salary_pred=pd.DataFrame(newdata,columns=['YearsExperience'])
```

```python
model.predict(salary_pred)
```

Hence, we can see, what LR model we build here is suitable for every new data point also So, it suitable model for given dataset.
